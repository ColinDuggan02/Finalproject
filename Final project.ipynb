## Question we were Interested in Addressing:
The internet has, especially in recent years, “added weight to utility status” because it is a leveler; it promotes social mobility, grants access to information and education as well as allows consumer’s access to cheaper goods and services, creating competition. It is indisputable that technological innovation correlates with economic growth, and “because of technological progress, daily life has changed more in the last 200 years than in the preceding 2000 years”. Technological progress is vital for economic growth since it allows productivity to increase, meaning the quantities of factors of production are now able to produce higher levels of output. Therefore, it seems obvious that technological advancement is an explanation for the increased productivity that can be seen across the world today. Since technology is nonrivalrous and oftentime nonexcludable due to it being mostly ideas and not tangible goods, it would seem intuitive that all countries would reap the benefits of technological innovation at very similar levels. This is not the case, however, similar to both physical and human capital, the countries which are at the cutting edge of technology are concentrated in HICs and account for only 14% of the world population. While HICs countries are able to easily innovate, transfer, imitate and use cutting edge technology, it is not easy to do so between HICs and LICs or even MICs. It is obvious that technological gaps exist between countries, therefore the data we researched aimed to answer: Would increased internet access and technological development in poor countries lead to sustainable economic growth and allow them to catch up with developing countries as well as what barriers exist that prevent fast technological transfer? Additionally, how has technology and more specifically internet access affected economic growth in the United States and in the rest of the World? Lastly, what role do tools like secure internet servers and patents protected by governments play in technological innovation?


## Assumptions:
Since private firms are primarily concerned with maximizing profits and raising the quality of their products, a massive part of R&D is done by private firms. It is important to keep this in mind since the pace of technological change will depend extensively on the decisions of private firms. Our data from the World Bank takes this into consideration. 
Technological Innovation that aids economic growth ranges from railroads and healthcare discoveries to microchips and AI. Spending on R&D encompasses all sectors but the rest of the data we analyze focuses on internet use, mobile cellular subscriptions and patents. 

## Definitions: 
### Improvement in Technology: 
Given quantities of physical and human capital can be combined to produce more output than previously possible.
### Patent: 
Grant made by a Government that confers on the creator of the invention the sole right to make, use and sell the invention for a period of time. 



## What we looked at:

### Expenditure on R&D as a percent of GDP
How much of total GDP is allocated towards research and development by country income groups. This data encompasses expenditures from four main sectors, Business enterprise, Higher Education, Private non-profit and Government. 
### GDP Growth (Change)
In order to see growth or stagnation by country income groups during the same years since data for R&D expenditure and internet usage as well as mobile phone subscriptions, and internet usage. 
### Individuals using the Internet as a percent of total Population
How many individuals have internet access and used the internet (in the last three months) via computer, phone, digital assistant, gaming systems etc. during the same years since data for R&D and GDP change. 
### Patent Applications (residents)
Patent applications filed through PCT in order to protect inventions of the owner for a period of time. An indicator for appetite in innovation.
### Secure Internet Servers
The number of distinct publicly trusted TLS/SSL certificates. An indicator for trust in using technology within country groups. 
### High Technology Exports as a percent of Manufactured Exports
These are exports with high R&D Intensity, i.e. aerospace, computers, scientific instruments, and electrical machinery. In order to see which countries are diffusing technological information.  
### Mobile Cellular Subscriptions
Subscriptions to a public mobile telephone service that provides access to the PSTN using cellular technology. Split into postpaid and prepaid subscriptions (in the last three months). In order to see basic infrastructure (power and communication lines) needed to support R&D investments as well as telecommunication investments. 




These seven Indicators are the basis for viewing which countries are the leaders in technological innovation and provide insight on the qualities that they all have in common as well as provide the  shared characteristics of the economies which are not technologically advanced. There are many indicators we could have chosen to assess “technological advancement”. We chose to look at Internet usage extensively along with the other indicators in order to achieve our goal of analyzing how internet usage (funded by government sponsored R&D) catalyzed technological innovation and provided an outlet for technological transfer.
